```python
# Reading whole data:
Master_data=pd.read_csv("/content/drive/MyDrive/Regression_2/Master_Data_mod_RUL.csv")
```

```python
# Doing train test split - 80% 20 % - and saving those files 
train, test = train_test_split(Master_data, test_size=0.2,random_state=3,shuffle=True,stratify=Master_data[["Engine_ID","Status", "Sub_Status"]])
```

```python
train.columns
```

```python
train_data1= train.drop(columns= ['Sub_Status', 'Unnamed: 0','Status'])
test_data1= test.drop(columns= ['Sub_Status', 'Unnamed: 0','Status'])
```

```python
train_data1
```

```python
X_train=train_data1.drop(['Engine_ID','RUL'], axis=1)
y_train= train_data1['RUL']

X_test=test_data1.drop(['Engine_ID','RUL'], axis=1)
y_test= test_data1['RUL']
```

```python
X_train
```

### Random Forest

```python
regr_RF = RandomForestRegressor(random_state=3)
regr_RF.fit(X_train, y_train)
y_pred_RF= regr_RF.predict(X_test)
print(regr_RF.score(X_test, y_test))
mean_absolute_error(y_test, y_pred_RF)
rmse_rf_val= np.sqrt(mean_squared_error(y_test, y_pred_RF)) 
print("Test RMSE is ",rmse_rf_val)
```

```python
def plot_rfscore_mae(df, id):
  my_df= df[df['Engine_ID']==str(id)]
  final_df = my_df.sort_values(by=['Cycle'], ascending=True)

  X_test= final_df.loc[:, ~final_df.columns.isin(['Engine_ID', 'RUL'])]
  y_test= final_df.iloc[:,-1]
  y_pred_rf= regr_RF.predict(X_test)
  #plotting
  plt.plot(final_df.Cycle, y_pred_rf, 'r') # plotting t, a separately 
  plt.plot(final_df.Cycle, y_test, 'b') # plotting t, b separately 
  plt.legend(["Predicted", "Original"], loc ="upper right")
  plt.show()

  return (regr_RF.score(X_test, y_test), mean_absolute_error(y_test, y_pred_rf), np.sqrt(mean_squared_error(y_test, y_pred_rf)))
```

```python
rf_score, mae_rf, rmse_rf= plot_rfscore_mae(test_data1, '1_2')
print(rf_score, mae_rf, rmse_rf)
```
